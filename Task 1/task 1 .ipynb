```python
# ===========================================================================
# TASK 1: IRIS DATASET EXPLORATION & VISUALIZATION
# ===========================================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Load the data first so 'df_iris' exists!
df_iris = pd.read_csv('Iris.csv')

print(f"Dataset Shape (Rows, Columns): {df_iris.shape}")
print(f"Column Headers: {list(df_iris.columns)}")
print("\nFirst 5 rows of the dataset:")
print(df_iris.head())
print("\n")



# Check for missing values
missing_values = df_iris.isnull().sum().sum()
print(f"Total missing values found: {missing_values}")
if 'Id' in df_iris.columns:
    df_iris = df_iris.drop(columns=['Id'])
    print("Dropped unnecessary 'Id' column.")
print("\n")


sns.set_theme(style="whitegrid")

plt.figure(figsize=(7, 4))
sns.scatterplot(data=df_iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species", palette="Set2")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()


plt.figure(figsize=(7, 4))
sns.histplot(data=df_iris, x="PetalLengthCm", hue="Species", kde=True, element="step")
plt.title("Histogram: Distribution of Petal Length across Species")
plt.xlabel("Petal Length (cm)")
plt.show()


plt.figure(figsize=(7, 4))
sns.boxplot(data=df_iris, x="Species", y="SepalWidthCm", hue="Species", palette="Pastel1", legend=False)
plt.title("Box Plot: Sepal Width Outlier Detection")
plt.ylabel("Sepal Width (cm)")
plt.show()

```
